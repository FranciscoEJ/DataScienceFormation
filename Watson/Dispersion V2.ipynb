```python
def cleaning(df):
    
    df=df.dropna(axis=1, how='all')
    sectores = df[' Descripción'].tolist()[2:]
    sectores = [x for x in sectores if (pd.isna(x) == False)]
    
    result = pd.DataFrame({'year':df.columns.tolist()[2:]})
    
    result.set_index('year',inplace=True)
    result.rename({'2007 R/':'2007', '2018 P/':'2018',
                   'Unnamed: 3':'2007-1','Unnamed: 4':'2007-2','Unnamed: 5':'2007-3',
                   'Unnamed: 8':'2008-1','Unnamed: 9':'2008-2','Unnamed: 10':'2008-3',
                  'Unnamed: 13':'2009-1','Unnamed: 14':'2009-2','Unnamed: 15':'2009-3',
                  'Unnamed: 18':'2010-1','Unnamed: 19':'2010-2','Unnamed: 20':'2010-3',
                   'Unnamed: 23':'2011-1','Unnamed: 24':'2011-2','Unnamed: 25':'2011-3',
                   'Unnamed: 28':'2012-1','Unnamed: 29':'2012-2','Unnamed: 30':'2012-3',
                   'Unnamed: 33':'2013-1','Unnamed: 34':'2013-2','Unnamed: 35':'2013-3',
                   'Unnamed: 38':'2014-1','Unnamed: 39':'2014-2','Unnamed: 40':'2014-3',
                   'Unnamed: 43':'2015-1','Unnamed: 44':'2015-2','Unnamed: 45':'2015-3',
                   'Unnamed: 48':'2016-1','Unnamed: 49':'2016-2','Unnamed: 50':'2016-3',
                   'Unnamed: 53':'2017-1','Unnamed: 54':'2017-2','Unnamed: 55':'2017-3',
                   'Unnamed: 58':'2018-1','Unnamed: 59':'2018-2','Unnamed: 60':'2018-3'
                  },axis=0,inplace=True)
    result.reset_index(inplace=True)
    df.set_index(' Descripción',inplace=True)

    for sector in sectores:
        result['{0}'.format(sector)] = df.loc[sector,:].tolist()[1:]

    return result



                   
```

```python
df_data =cleaning(df_data_1)
```

```python
def number(x):
    x = x.split(' ')
    return float(''.join(x))
```

```python
for col in df_data.columns.tolist()[1:]:
    df_data[col] = df_data[col].apply(lambda x: number(x))
```
